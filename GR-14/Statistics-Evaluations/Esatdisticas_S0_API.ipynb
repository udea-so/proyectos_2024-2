```python
# Paso 1: Importar las bibliotecas necesarias
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


# Paso 3: Leer los datos monitoreo y de perfilado
data_path = 'datos_monitoreo.csv'
profiling_path = 'datos_perfilado.csv'

df = pd.read_csv(data_path)
profiling_data = pd.read_csv(profiling_path)

# Paso 4: Crear gr치ficos
```

Uso promedio de CPU por escenario y equipo

```python
# Gr치fico 1: Uso promedio de CPU por escenario y equipo
plt.figure(figsize=(10, 6))
sns.barplot(data=df, x='Escenario', y='Uso_CPU', hue='Equipo', ci='sd')
plt.title('Uso promedio de CPU por escenario y equipo', fontsize=16)
plt.ylabel('Uso de CPU (%)')
plt.xlabel('Escenario')
plt.legend(title='Equipo')
plt.tight_layout()
plt.show()
```

Distribuci칩n del uso de memoria

```python
# Gr치fico 2: Distribuci칩n del uso de memoria por escenario y equipo
plt.figure(figsize=(10, 6))
sns.violinplot(data=df, x='Escenario', y='Uso_Memoria', hue='Equipo', split=True)
plt.title('Distribuci칩n del uso de memoria por escenario y equipo', fontsize=16)
plt.ylabel('Uso de Memoria (%)')
plt.xlabel('Escenario')
plt.legend(title='Equipo')
plt.tight_layout()
plt.show()
```

 Desempe침o del perfilado (tiempo promedio por llamada)

```python
# Gr치fico 3: Tiempo promedio por llamada en el perfilado
plt.figure(figsize=(8, 5))
sns.barplot(data=profiling_data, x='Equipo', y='Tiempo Promedio (ms)', palette='viridis')
plt.title('Tiempo promedio por llamada en el perfilado', fontsize=16)
plt.ylabel('Tiempo Promedio (ms)')
plt.xlabel('Equipo')
plt.tight_layout()
plt.show()
```
